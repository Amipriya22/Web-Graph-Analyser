```python
import pandas as pd
import numpy as np  
import re
```

```python
# i have 3 excel files contianing web crwled data, what i want is , in each of the files there many coloumns but i have to specifically work with to and from coloumn, so i want to make a seperate excel file continaing only the 2 coloumn from and to, whose vvalues are taken from the from and to coloumn sof the 3 excel files, so basicly i have to create 2 single columns from and to that have values from to and from column of 3 excel files 
df1 = pd.read_excel('wiki_iitk_step1.xlsx')
df2 = pd.read_excel('outlinks_seed_part1.xlsx')
df3 = pd.read_excel('outlinks_seed_part2.xlsx')
```

```python
print(df1.shape)
print(df2.shape)
print(df3.shape)
print(df1.shape[0]+df2.shape[0]+df3.shape[0])
```

```python
df_from = pd.concat([df1['From'], df2['From'], df3['From']], ignore_index=True)
df_to = pd.concat([df1['To'], df2['To'], df3['To']], ignore_index=True)
df_final = pd.DataFrame({'From': df_from, 'To': df_to})
```

```python
df_final.shape
```

```python
df_final.head(20)
```

```python

# Define a cleaning function
def clean_url(url):
    if pd.isna(url):
        return ""
    # Remove protocol and www
    url = re.sub(r"^https?://(www\.)?", "", url)
    # Remove trailing slash
    # url = url.rstrip("/")
    return url

# Apply to both columns
df_final["From"] = df_final["From"].apply(clean_url)
df_final["To"] = df_final["To"].apply(clean_url)
df_final.head(20)
```

```python
df_final.tail(20)
```

```python
# Filter out extremely long addresses (>100 characters)
df_final = df_final[df_final["From"].str.len() <= 100]
df_final = df_final[df_final["To"].str.len() <= 100]
print(df_final.shape)
```

```python

# Optional: Remove links with noisy patterns
noise_keywords = ["app", "payment", "redirect", "utm_", "sessionid", "checkout"]
pattern = "|".join(noise_keywords)

df_final = df_final[~df_final["From"].str.contains(pattern, case=False, na=False)]
df_final = df_final[~df_final["To"].str.contains(pattern, case=False, na=False)]

df_final.shape
```

```python
# Optional: Remove duplicate edges
df_final = df_final.drop_duplicates(subset=["From", "To"])
df_final.shape
```

```python
# Save cleaned version
df_final.to_csv("cleaned_links1.csv", index=False)
```
